```python
import pandas as pd
import openpyxl

pd.set_option('display.max_rows', 100)
```

```python
# Import data
df_first = pd.read_excel('./Names2020_FirstNames_RaceHispanic.xlsx', skiprows=2)
df_first = df_first[:-1]

df_last = pd.read_excel('./Names2020_LastNames_RaceHispanic.xlsx', skiprows=2)
df_last = df_last[:-1]
```

```python
# Sort by first name popularity
df_asian_first = df_first.sort_values(by=df_first.columns[8], ascending=False)
df_hispanic_first = df_first.sort_values(by=df_first.columns[-1], ascending=False)
df_native_first = df_first.sort_values(by=df_first.columns[7], ascending=False)
```

```python
# Sort by last name popularity
df_asian_last = df_last.sort_values(by=df_last.columns[8], ascending=False)
df_hispanic_last = df_last.sort_values(by=df_last.columns[-1], ascending=False)
df_native_last = df_last.sort_values(by=df_last.columns[7], ascending=False)
```
